```python
df = pd.read_csv('bip-2021.csv', index_col=0)
```

```python
pd.set_option('display.max_columns', None)
df
```

```python
plt.scatter(df.launch_speed, df.launch_angle, s=5, c=df.xBA, cmap='YlOrRd')
plt.xlabel('Exit Velocity (MPH)')
plt.ylabel('Launch Angle')
plt.title(label='Hit Probability (xBA)', loc='left', fontsize=17, y=1.18)
plt.suptitle(t='The launch angle and exit velocity combinations of every batted\nball in play in the 2021 season. Light-colored points are more likely\nto result in outs. Darker points are more likely to result in hits.', ha='left', y=1.01, x=.125, fontsize=10)
plt.axvline(x=95, color='gray', linestyle='-')

cax = plt.axes([.93, 0.3, 0.05, 0.3])
cbar = plt.colorbar(cax=cax)
cbar.set_label('xBA', labelpad=-40, y=1.22, rotation=0)
cbar.ax.set_yticklabels(['.000', '.250', '.500', '.750', '1.000'])
plt.savefig('xba-chart.svg', bbox_inches='tight')
```
